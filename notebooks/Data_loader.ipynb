```python
import pandas as pd

# Load the data
df = pd.read_csv('merged crime data fixed.csv')

# Print the head of the dataframe
print("Head of the data:")
print(df.head())

# Print the column names
print("\nColumn names:")
print(df.columns.tolist())

# Print unique values for the 'crime' column (assuming the column is named 'crime')
if 'crime' in df.columns:
    print("\nUnique values in 'crime' column:")
    print(df['crime'].unique())
else:
    print("\nNo 'crime' column found in the data.")

# Print the size of the data
print(f"\nShape of the data: {df.shape}")

# Print data types and non-null counts
print("\nData info:")
print(df.info())

# Print summary statistics for numeric columns
print("\nSummary statistics:")
print(df.describe())
```
